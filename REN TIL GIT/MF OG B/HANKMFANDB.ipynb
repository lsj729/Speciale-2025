```python
import numpy as np
import pickle

IRF_KEYS = ["C_hh", "A_hh", "ra", "Z", "chi", "pi", "Y", "i", "r"]

def _get_irfs(model, keys=IRF_KEYS):
    out = {}
    for k in keys:
        if k not in model.IRF:
            raise KeyError(f"Model missing IRF key '{k}'")
        out[k] = np.asarray(model.IRF[k]).ravel()
    return out

def input_contrib_series(model, inp, output="C_hh", T=300, scale="pct"):
    key = (output, inp)
    if key not in model.jac_hh:
        raise KeyError(f"Missing Jacobian key {key}")
    if inp not in model.IRF:
        raise KeyError(f"Missing IRF path for '{inp}'")

    J = np.asarray(model.jac_hh[key])[:T, :T]
    x = np.asarray(model.IRF[inp]).ravel()
    if x.size < T:
        x = np.pad(x, (0, T - x.size))
    else:
        x = x[:T]

    d = J @ x  # level contribution
    if scale == "pct":
        ssC = float(getattr(model.ss, output))
        d = 100.0 * d / ssC
    return d
```

```python
import numpy as np
import pickle

IRF_KEYS = ["C_hh", "A_hh", "ra", "Z", "chi", "pi", "Y", "i", "r"]

# Needed for plot_input_effect_on_C (aggregate Jacobians)
JAC_KEYS = [("C_hh","ra"), ("C_hh","Z"), ("C_hh","chi"),
            ("A_hh","ra"), ("A_hh","Z"), ("A_hh","chi")]

# Needed for table_C_impact_by_z_from_perz_decomp / decomposition (per-z Jacobians)
JAC_Z_KEYS = [("C_hh","ra"), ("C_hh","Z"), ("C_hh","chi"),
              ("A_hh","ra"), ("A_hh","Z"), ("A_hh","chi")]  # include A_hh if your decomp uses it


def _get_irfs(model, keys=IRF_KEYS):
    out = {}
    for k in keys:
        if k in model.IRF:
            out[k] = np.asarray(model.IRF[k]).ravel()
        else:
            raise KeyError(f"{getattr(model,'name','model')} missing IRF key '{k}'")
    return out


def _get_jacs(model, keys=JAC_KEYS, T=300):
    if not hasattr(model, "jac_hh"):
        raise AttributeError("Model has no jac_hh. Run compute_jacs(...) first.")
    out = {}
    for key in keys:
        if key in model.jac_hh:
            out[f"{key[0]}__{key[1]}"] = np.asarray(model.jac_hh[key])[:T, :T]
        else:
            # keep explicit missingness (helps debugging)
            out[f"{key[0]}__{key[1]}"] = None
    return out


def _get_jacs_z(model, keys=JAC_Z_KEYS, T=300):
    if not hasattr(model, "jac_hh_z"):
        raise AttributeError("Model has no jac_hh_z. Run compute_jac_hh_z(...) first.")
    out = {}
    for key in keys:
        if key in model.jac_hh_z:
            Jz = np.asarray(model.jac_hh_z[key])
            # expected shape: (Nz, T, T)
            out[f"{key[0]}__{key[1]}"] = Jz[:, :T, :T]
        else:
            out[f"{key[0]}__{key[1]}"] = None
    return out


def _get_ss_arrays(model):
    ss = model.ss
    out = {}

    # required by table_C_impact_by_z_from_perz_decomp
    if not hasattr(ss, "D"):
        raise AttributeError("ss.D missing (needed for mass(z)).")
    if not hasattr(ss, "c"):
        raise AttributeError("ss.c missing (needed for C_ss(z)).")

    out["D"] = np.asarray(ss.D)
    out["c"] = np.asarray(ss.c)

    # often useful / sometimes needed elsewhere
    for k in ["C_hh", "A_hh", "B", "pD", "Y", "Z", "chi", "ra", "r", "pi", "i"]:
        if hasattr(ss, k):
            val = getattr(ss, k)
            if np.isscalar(val):
                out[k] = float(val)

    return out


def _get_par_scalars(model):
    par = model.par
    out = {}
    for k in ["T", "beta", "mu", "rho_z", "sigma_psi", "phi_pi", "rho_epsilon_i"]:
        if hasattr(par, k):
            out[k] = float(getattr(par, k))
    return out


# ---- your three models in the other notebook ----
models = {
    "other_baseline": model_B,
    "other_sticky":   model_sticky_B,
    "other_hetero":   model_B_het,
}

Tsave = 300
contrib_inputs = ["ra", "Z", "chi"]

payload = {
    "IRF_KEYS": IRF_KEYS,
    "T": Tsave,
    "models": {},
}

for name, m in models.items():
    payload["models"][name] = {
        "name": getattr(m, "name", name),
        "IRF": _get_irfs(m, IRF_KEYS),

        # for plot_input_effect_on_C
        "jac_hh": _get_jacs(m, JAC_KEYS, T=Tsave),

        # for per-z decomposition/table
        "jac_hh_z": _get_jacs_z(m, JAC_Z_KEYS, T=Tsave),

        # steady-state arrays needed for mass(z) and C_ss(z)
        "ss": _get_ss_arrays(m),

        # some parameters (mainly T; harmless to include)
        "par": _get_par_scalars(m),

        # optional: you already computed these; still useful for quick plots
        "contrib": {
            inp: input_contrib_series(m, inp=inp, output="C_hh", T=Tsave, scale="pct")
            for inp in contrib_inputs
        },
    }

with open("other_notebook_bundle.pkl", "wb") as f:
    pickle.dump(payload, f)

print("Saved other_notebook_bundle.pkl with IRFs, jac_hh, jac_hh_z, and ss.D/ss.c.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_key_irfs_clean_up_to_three_models(
    model1,
    model2=None,
    model3=None,
    T=40,
    extra=("pi","Y","i","r"),
    title="IRFs",
    ylabel="Deviation from steady state",
    ncols=3,
    figsize_per_col=4.2,
    figsize_per_row=2.8,
    colors=("forestgreen", "navy", "red"),
    labels=("model 1", "model 2", "model 3"),
    ls=("-", "--", "-."),
    force_union=True,   # union of available vars across provided models
):
    """
    Plot key IRFs for 1, 2, or 3 models (overlayed in each subplot).

    Parameters
    ----------
    model1 : required, must have .IRF dict
    model2, model3 : optional, can be None
    force_union : if True, plot variables available in ANY provided model
                  if False, plot only variables available in ALL provided models
    """

    wanted = ["C_hh","A_hh","ra","Z","chi"] + list(extra)

    models = [m for m in (model1, model2, model3) if m is not None]
    if len(models) == 0:
        raise ValueError("At least one model must be provided.")

    # clip styles to number of models
    colors = list(colors)[:len(models)]
    labels = list(labels)[:len(models)]
    ls     = list(ls)[:len(models)]

    def _avail(m):
        out = []
        if not hasattr(m, "IRF"):
            return out
        for v in wanted:
            if v in m.IRF:
                arr = np.asarray(m.IRF[v])
                if np.any(np.isfinite(arr)):
                    out.append(v)
        return out

    avails = [set(_avail(m)) for m in models]

    if force_union:
        available = [v for v in wanted if any(v in a for a in avails)]
    else:
        available = [v for v in wanted if all(v in a for a in avails)]

    if len(available) == 0:
        raise RuntimeError("No requested IRF series found in provided model.IRF dicts.")

    n = len(available)
    nrows = int(np.ceil(n / ncols))

    fig, axes = plt.subplots(
        nrows, ncols,
        figsize=(figsize_per_col * ncols, figsize_per_row * nrows),
        squeeze=False
    )

    # ---- margins ----
    left, right, bottom, top = 0.12, 0.98, 0.10, 0.88
    wspace, hspace = 0.30, 0.55
    fig.subplots_adjust(left=left, right=right, bottom=bottom, top=top,
                        wspace=wspace, hspace=hspace)

    fig.text(0.5, 0.94, title, ha="center", va="center", fontsize=14)
    fig.text(0.04, 0.5, ylabel, ha="center", va="center", rotation="vertical", fontsize=12)

    def _get_series(m, var):
        if (not hasattr(m, "IRF")) or (var not in m.IRF):
            return None
        s = np.asarray(m.IRF[var]).ravel()
        if s.size == 0 or not np.any(np.isfinite(s)):
            return None
        return s[:T]

    # plot panels
    for idx, var in enumerate(available):
        r = idx // ncols
        c = idx % ncols
        ax = axes[r, c]

        series_list = [_get_series(m, var) for m in models]

        # plot each provided model if series exists
        for j, s in enumerate(series_list):
            if s is None:
                continue
            ax.plot(np.arange(s.size), s, lw=2.3, color=colors[j], ls=ls[j], label=labels[j])

        ax.axhline(0, color="black", lw=1)
        ax.set_title(var, fontsize=12)
        ax.set_xlabel("quarters")
        ax.grid(alpha=0.25)
        ax.set_ylabel("")

        # legend only once (first panel)
        if idx == 0 and len(models) > 1:
            ax.legend(frameon=True, fontsize=9)

    # turn off unused axes
    for idx in range(n, nrows * ncols):
        r = idx // ncols
        c = idx % ncols
        axes[r, c].axis("off")

    plt.show()
    return available
```
